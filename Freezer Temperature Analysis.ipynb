# Missing Value Imputation

```python
newdf = raw_data.fillna(raw_data.median()) # by default df.median() is computed along rows (axis=0) if not mentioned 
newdf.info()                               # so it will compute median of each column and fill NaN with median.
```

Here we can see that total rows in dataframe is 23040 and every column have 23040 non null value means missing value has been imputed successfully. If someone decides to impute mean then it can be done:
newdf = raw_data.fillna(raw_data.mean())


# Daily Average Temperature in Every Freezer

```python
#df = pd.read_csv("Temperature_Data.csv")
df = raw_data.reset_index()
newdf = pd.melt(df, id_vars="0", var_name="Site_Freezer", value_name="Temp")
newdf = newdf.sort_values("0").reset_index().drop("index", axis=1)
newdf.columns=["DateTime","Site_Freezer", "Temp"]
newdf["Site"]=newdf["Site_Freezer"].str.split(">").apply(lambda x:x[0])
newdf["Freezer"]=newdf["Site_Freezer"].str.split(">").apply(lambda x:x[1])
newdf.drop("Site_Freezer", axis=1, inplace=True)
newdf["DateTime"] = newdf.DateTime.astype(np.datetime64)
newdf.set_index("DateTime", inplace=True)
g = newdf.groupby(["Site","Freezer"])
resampledf = g["Temp"].resample("D").mean()
resampledf= pd.DataFrame(resampledf)
resampledf.head()
```

# Average Temperature at Every Site

```python
plotdf = resampledf.reset_index()
fig, ax = plt.subplots(1,1,figsize=(20,20))
ax = sns.lineplot(x="DateTime", y="Temp", hue = "Site" ,data =plotdf)
start, end = ax.get_xlim()
ax.yaxis.set_ticks(np.arange(-20, 25, 1))
plt.xticks(rotation=45)
plt.xlabel("Date")
plt.ylabel("Average daily Temperature")
plt.show()
```

# Average Temperature in Every Freezer at Site-1

```python
plotdf = resampledf.reset_index()
plotdf2 = plotdf[plotdf.Site=="Site-1 "]
fig, ax = plt.subplots(1,1,figsize=(20,20))
ax = sns.lineplot(x="DateTime", y="Temp", hue = "Freezer" ,data =plotdf)
start, end = ax.get_xlim()
ax.yaxis.set_ticks(np.arange(-20, 25, 1))
plt.xticks(rotation=45)
plt.xlabel("Date")
plt.ylabel("Average daily Temperature")
plt.show()
```
